```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

### Define model

```python
# Here we are using 2 diffusion models
model_1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_2 = "stabilityai/stable-diffusion-xl-base-1.0"
```

```python
from diffusers.utils.import_utils import USE_SAFETENSORS
# pipeline
pipe_1 = StableDiffusionPipeline.from_pretrained(model_1, dtype=torch.float16,use_safetensors=True)
pipe_1 = pipe_1.to("cuda")
```

```python
prompt = """dreamlikeart, a grungy woman with rainbow hair, travelling between dimensions, dynamic pose, happy, soft eyes and narrow chin,
extreme bokeh, dainty figure, long hair straight down, torn kawaii shirt and baggy jeans
"""
```

```python
image = pipe_1(prompt=prompt).images[0]
image
```

```python
plt.imshow(image)
plt.axis('off')
```

```python
prompt2 = """A girl is sittig on a chair & She is accompanied by her tiger. Make sure to keep it cinematic and color to be golden iris
"""

image = pipe_1(prompt2).images[0]
image
plt.imshow(image)
plt.axis('off')
```
